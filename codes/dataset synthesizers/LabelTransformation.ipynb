```python
samples_name = "SyntheticArrows"

samples_zip = f"{samples_name}_test_samples.zip"
samples_folder = f"{samples_name}_test_samples"
sample_url = f"path/to/sample/{samples_name}_test_samples.zip"

!wget $sample_url
!unzip $samples_zip
```

```python
import os
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
data_root = samples_folder
sample_names = sorted(os.listdir(data_root))

!rm -r gt_labels
!mkdir gt_labels

for sample_name in sample_names:
  with open(os.path.join('./gt_labels', f"{sample_name}_gt.txt"), "w") as out_file:
    out_file.write("F\tCELLNUM\tX\tY\n")
    for frame in range(100):
      label = np.array(Image.open(os.path.join(data_root, sample_name, 'labels', f'{sample_name}_{frame:03d}_label.png')), dtype='int')
      IDs = sorted(np.unique(label))
      for ID in IDs:
        if ID !=0:
          single_label = np.array(label == ID, dtype='uint8')
          contours, hierarchy  = cv2.findContours(single_label ,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
          for contour in contours:
              for a in contour:
                l=a[0]
                out_file.write(str(frame+1)+"\t"+str(ID)+"\t"+str(l[0])+"\t"+str(l[1])+"\n")
  print(f"Sample {sample_name} label transformed")
```
